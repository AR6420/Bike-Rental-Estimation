#### <font color='blue'> Imputing missing values in the column Functioning Day. It is a Binomial Column and NaN values are replace by Highest Frequency, Then match the Highest frequency to 1 and other to 0 </font>

```python
traindata['Functioning Day'].fillna('Yes',inplace=True)
testdata['Functioning Day'].fillna('Yes',inplace=True)
## mapping Yes to 1 and No to 0
traindata['Functioning Day'].replace(['Yes','No'],[1,0],inplace=True)
testdata['Functioning Day'].replace(['Yes','No'],[1,0],inplace=True)
```

#### <font color='blue'>Imputing missing values in the column "holiday" to the most frequent element since it is binomial data, then mapping no holiday with 1 and no holiday with 0</font>

```python
## fill the NaN with No Holiday in column Hoilday in train and test data
traindata['Holiday'].fillna('No Holiday',inplace=True)
testdata['Holiday'].fillna('No Holiday',inplace=True)
## mapping No Holiday to 1 and Holiday to 0 in both the dataset
traindata['Holiday'].replace(['No Holiday','Holiday'],[1,0],inplace=True)
testdata['Holiday'].replace(['No Holiday','Holiday'],[1,0],inplace=True)
```

#### <font color='blue'> Handling missing data in the snowfall (cm) and Rainfall(mm) columns.The snowfall and Rainfall median value is zero across all seasons, so we are imputing missing values in the snowfall (cm) column in train and test with 0.0</font>  

```python
### filling NaN in Snowfall (cm) with 0.0 in the train and test dataset
traindata['Snowfall (cm)'].fillna(0.0,inplace=True)
testdata['Snowfall (cm)'].fillna(0.0,inplace=True)
### filling NaN in Rainfall(mm) with 0.0 in the train and test dataset
traindata['Rainfall(mm)'].fillna(0.0,inplace=True)
testdata['Rainfall(mm)'].fillna(0.0,inplace=True)
```

#### <font color='blue'>While handling missing values in the column Solar Radiation (MJ/m2), the solar radiation performance across various seasons and timelines is observed. From the solar radiation density graph, we could conclude that it is the nature of the normal curve and we are replacing the missing values in solar radiation by the mean of its timeline and season.In the case of NaN in independent column, we are imputing it with the overal mean. Because solar radiation is important to see the timeline, by imputing everything with the mean, the data set's nature is distorted.</font>

#### <font color='green'> Filling the weather values based on the season's analysis</font>

```python
### solar radiation mean across the dataset 
solaravg=round(traindata['Solar Radiation (MJ/m2)'].mean(),2)
```

```python
### solar radiation mean aross the seasons and timline
solarval=traindata.groupby(['Seasons','TimeLine'])['Solar Radiation (MJ/m2)'].mean().values.round(2).tolist()
```

```python
### Observing the mean value group by season and timeline
traindata.groupby(['Seasons','TimeLine'])['Solar Radiation (MJ/m2)'].mean()
```

```python
### function to impute the missing value in the solar radiation column
def impute_solar(cols):
    solar= cols[0]
    season= cols[1]
    time=cols[2]
    
    if (pd.isnull(season) or pd.isnull(time)) and pd.isnull(solar): return solaravg
    
    elif pd.isnull(solar):
        if season=='Autumn':
            if time=='NoLight':return  solarval[0]
            if time=='Office':return solarval[1]
            if time=='Routine':return solarval[2]
            if time=='Transits':return solarval[3]
     
        if season=='Spring':
            if time=='NoLight':return  solarval[4]
            if time=='Office':return solarval[5]
            if time=='Routine':return solarval[6]
            if time=='Transits':return solarval[7]
        if season=='Summer':
            if time=='NoLight':return  solarval[8]
            if time=='Office':return solarval[9]
            if time=='Routine':return solarval[10]
            if time=='Transits':return solarval[11]
       
        if season=='Winter':
            if time=='NoLight':return  solarval[12]
            if time=='Office':return solarval[13]
            if time=='Routine':return solarval[14]
            if time=='Transits':return solarval[15]
    else:
        return solar
```

```python
### observing some common null values with Solar Radiation (MJ/m2)
traindata[traindata['Seasons'].isnull()&traindata['Solar Radiation (MJ/m2)'].isnull()]
```

```python
### observing some common null values with Solar Radiation (MJ/m2)
traindata[traindata['TimeLine'].isnull()&traindata['Solar Radiation (MJ/m2)'].isnull()]
```

```python
### imputing missing values in the train dataset
traindata['Solar Radiation (MJ/m2)'] = traindata[['Solar Radiation (MJ/m2)','Seasons','TimeLine']].apply(impute_solar,axis=1)
### imputing missing values in the testdataset
testdata['Solar Radiation (MJ/m2)'] = testdata[['Solar Radiation (MJ/m2)','Seasons','TimeLine']].apply(impute_solar,axis=1)
```

### <font color='blue'> Applying K Means to the data frame, removes column seasons,timeline and rentedbike count from dataframe, scale it and cluster it after clustering refill the NaN values </font>

```python
# import libraries for clustering and declaring the object kmeans 
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
kmeans = KMeans(n_clusters = 4,  random_state = 0)
## declaring the scaler object 
scaler = MinMaxScaler()
## fit the scaler model accordin to the train data set
scaler.fit(traindata.drop(['Seasons','TimeLine','Rented Bike Count'],axis=1))
```

```python
## using the standard scaler object created and scaling the dataset
Std_clust_train=scaler.transform(traindata.drop(['Seasons','TimeLine','Rented Bike Count'],axis=1))
Std_clust_test=scaler.transform(testdata.drop(['Seasons','TimeLine'],axis=1))
```
